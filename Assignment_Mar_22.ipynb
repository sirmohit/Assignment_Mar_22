Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Ans:
    
    The Pearson correlation coefficient, often denoted as "r," is a statistical measure that quantifies the strength and 
    direction of a linear relationship between two variables. It ranges from -1 to +1, where:

    +1 indicates a perfect positive linear relationship,
    -1 indicates a perfect negative linear relationship,
    0 indicates no linear relationship.
    To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and 
    their final exam scores, you can use the following formula:
        r = ∑(Xi− Xmean)(yi- ymean)/root of  ∑(Xi− Xmean)^2 *  ∑(Yi− Ymean)^2 
    Where:
     Xi and Yi are the individual data points for the two variables,
    ˉXand ˉY are the means of the two variables.
    Once you calculate the correlation coefficient:

    If r is close to +1, it suggests a strong positive linear relationship. This would mean that as the amount of time spent 
    studying increases, the final exam scores tend to increase as well.

    If r is close to -1, it indicates a strong negative linear relationship. This would imply that as the time spent studying 
    increases, the final exam scores tend to decrease.

    If r is close to 0, there is little to no linear relationship between the two variables. Studying time and exam scores do
    not show a strong linear trend.

    Interpreting the result depends on the actual value of
    r and whether it is statistically significant. Additionally, it's essential to keep in mind that correlation does not
    imply causation; a correlation between two variables does not necessarily mean that one causes the other.








Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

    Ans:

        Spearman's rank correlation coefficient (ρ) assesses the strength and direction of a monotonic relationship 
        between two variables. Unlike the Pearson correlation,
    Spearman's correlation does not assume a linear relationship; it looks at whether there is a consistent upward or 
    downward trend in the ranks of the data points.

    To calculate Spearman's rank correlation:

    Rank the data for each variable separately.
    Assign ranks, with the smallest observation receiving a rank of 1, the next smallest a rank of 2, and so on.
    If there are tied values, assign them the average of the ranks they would have received.
    Calculate the differences between the ranks for each pair of corresponding data points.
    Square these differences.
    Sum up the squared differences.
    Use the formula for Spearman's rank correlation:
    ρ=6∑di^2/n(n^2 - 1)
    Where:

    di is the difference between the ranks of corresponding data points.

    n is the number of data points.
    The resulting 
    ρ value will range from -1 to +1, where:
    ρ=+1 indicates a perfect monotonic increasing relationship.
    ρ=−1 indicates a perfect monotonic decreasing relationship.
    ρ=0 indicates no monotonic relationship.
    Interpreting the result:

    If ρ is close to +1, it suggests a strong monotonic increasing relationship. This would mean that as the amount of
    sleep individuals get each night increases, their job satisfaction level tends to increase.

    If ρ is close to -1, it indicates a strong monotonic decreasing relationship. This would imply that as the amount of 
    sleep decreases, job satisfaction tends to decrease.

    If ρ is close to 0, there is little to no monotonic relationship between the two variables. Sleep duration and job 
    satisfaction do not show a strong consistent trend.

    Remember that Spearman's rank correlation assesses monotonicity, not necessarily linearity, and it is robust to outliers.

    Pearson Correlation (r):

    Assumes a linear relationship between variables.
    Sensitive to outliers.
    Measures the strength and direction of a linear relationship.
    Ranges from -1 to +1.
    Spearman's Rank Correlation (ρ):

    Does not assume a linear relationship; it assesses monotonicity.
    Less sensitive to outliers.
    Measures the strength and direction of a monotonic relationship.
    Ranges from -1 to +1.
    Now, let's assume you have the data and calculate both correlation coefficients. After ranking the data for the
    Spearman's correlation, you can compute the two coefficients and compare the results.

    If r is close to -1, it indicates a strong negative linear relationship. If ρ is close to -1, it indicates a strong 
    monotonic decreasing relationship.

Q4. A researcher is interested in examining the relationship between the number of hours individuals

both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.

    To calculate the Pearson correlation coefficient (r) between the number of hours individuals spend watching television 
    per day and their level of physical activity, you can use the following formula:
          r = ∑(Xi− Xmean)(yi- ymean)/root of  ∑(Xi− Xmean)^2 *  ∑(Yi− Ymean)^2 
    Where:
     Xi and Yi are the individual data points for the two variables,
    ˉXand ˉY are the means of the two variables.
    
    This formula essentially computes the covariance of the two variables divided by the product of their standard 
    deviations. Here's a step-by-step guide:

    Calculate the mean (X) and standard deviation (Sx) for the number of hours spent watching television.
    Calculate the mean (Y) and standard deviation (Sy) for the level of physical activity.
    For each participant, calculate (Xi -X) and (Yi - Y) for both variables.
    Multiply these differences for each participant:(Xi -X)(Yi - Y)
    Calculate:root of  ∑(Xi− Xmean)^2 and  ∑(Yi− Ymean)^2 
    Divide the sum of products by the product of the standard deviations: 
         r = ∑(Xi− Xmean)(yi- ymean)/root of  ∑(Xi− Xmean)^2 *  ∑(Yi− Ymean)^2 
    The resulting r value will be between -1 and +1, indicating the strength and direction of the linear relationship:

    If r is close to +1, it suggests a strong positive linear relationship: as the number of hours spent watching 
        television increases, the level of physical activity tends to increase.

    If r is close to -1, it indicates a strong negative linear relationship: as the number of hours spent watching 
        television increases, the level of physical activity tends to decrease.

Ans:
    
    To examine the relationship between age and soft drink preference, you might want to use descriptive statistics and 
    possibly some graphical representation. However, it's important to note that the data you've provided doesn't have 
    a clear quantitative relationship between the variables. If you have more data or specific preferences associated 
    with each age, you could explore the relationship further.

    Here are some basic descriptive statistics and visualizations you could use with the given data:

    Descriptive Statistics for Age:

    Mean (average) age
    Median (middle value) age
    Mode (most frequently occurring) age
    Soft Drink Preference Frequency:

    Count the frequency of each soft drink preference (Coke, Pepsi, Mountain Dew).
    Bar Chart:

    Create a bar chart to visually represent the frequency of each soft drink preference.




```python
# Example
import matplotlib.pyplot as plt
import pandas as pd

# Provided data
age_data = [25, 42, 37, 19, 31, 28]
preference_data = ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']

# Create a DataFrame
df = pd.DataFrame({'Age': age_data, 'Preference': preference_data})

# Descriptive statistics for Age
print("Descriptive Statistics for Age:")
print(df['Age'].describe())
print("\n")

# Soft Drink Preference Frequency
preference_counts = df['Preference'].value_counts()
print("Soft Drink Preference Frequency:")
print(preference_counts)
print("\n")

# Bar Chart
plt.bar(preference_counts.index, preference_counts)
plt.title("Soft Drink Preference")
plt.xlabel("Soft Drink")
plt.ylabel("Frequency")
plt.show()
```
